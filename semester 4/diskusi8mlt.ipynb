```python
import pandas as pd
import numpy as np
```

```python
# Define covariance matrices S1 and S2
S1 = np.array([[13825.3, 23823.4],
               [23823.4, 73107.4]])

S2 = np.array([[8632, 19616.7],
               [19616.7, 55964.5]])

print("S1:")
print(S1)
print("\nS2:")
print(S2)
```

```python
n1=77
n2=77
p=2
```

```python
(n1-1)+(n2-1)
```

```python
S=(1/((n1-1)+(n2-1)))*((n1-1)*(S1)+(n2-1)*(S2))
```

```python
S
```

```python
det_S = np.linalg.det(S)
det_S1 = np.linalg.det(S1)
det_S2 = np.linalg.det(S2)
```

```python
print("Determinant of S:", det_S)
print("Determinant of S1:", det_S1)
print("Determinant of S2:", det_S2)
```

```python
M=((n1-1)+(n2-1))*np.log(det_S)-(n1-1)*np.log(det_S1)-(n2-1)*np.log(det_S2)
M
```

```python
k=2
```

```python
1/(n1-1)+1/(n2-1)
```

```python
1/152
```

```python
1/((n1-1)+(n2-1))
```

```python
C_ = 1 - (((2*p**2 + 3*p -1)/(6*(p+1)*(k-1)))*(1/(n1-1)+1/(n2-1)-1/((n1-1)+(n2-1))))
```

```python
C_
```

```python
MC=M*C_
```

```python
MC
```

```python
df=p*(p+1)*(k-1)/2
df
```

```python
from scipy.stats import chi2

titik_kritis = chi2.ppf(0.90, df)
titik_kritis
```

```python
print("kesimpulan: ")
if MC>titik_kritis:
    print("Tolak H0, berarti ada perbedaan yang signifikan antara kedua varians")
else:
    print("Gagal tolak H0, berarti tidak ada perbedaan yang signifikan antara kedua varians")
```
