```python
import pandas as pd
import numpy as np
```

```python
m1 = np.array([[6, 7]])
m2 = np.array([[5, 9]])
m3 = np.array([[4, 6]])
m4 = np.array([[6, 6]])
m5 = np.array([[4, 7]])
m6 = np.array([[5, 4]])
m7 = np.array([[6, 4]])

```

```python
df = pd.DataFrame({
    'Y1': [6, 5, 4, 6, 4, 5, 6],
    'Y2': [7, 9, 6, 6, 7, 4, 4],
    'Perlakuan': [1, 1, 2, 2, 2, 3, 3]
})
```

```python
df
```

# SOAL 1

```python
m_rata2_p1 = (m1 + m2) / 2
m_rata2_p2 = (m3 + m4 + m5) / 3
m_rata2_p3 = (m6 + m7) / 2
m_rata_all = (m1 + m2 + m3 + m4 + m5 + m6 + m7) / 7
```

```python
print("m_rata2_p1:")
print(m_rata2_p1)
print("\nm_rata2_p2:")
print(m_rata2_p2)
print("\nm_rata2_p3:")
print(m_rata2_p3)
print("\nm_rata_all:")
print(m_rata_all)
```

# SOAL 3

## A

```python
selisih_p1 = m_rata2_p1 - m_rata_all
selisih_p2 = m_rata2_p2 - m_rata_all
selisih_p3 = m_rata2_p3 - m_rata_all
```

```python
A1=len(df['Perlakuan'][df['Perlakuan']==1])*selisih_p1*selisih_p1.T
A2=len(df['Perlakuan'][df['Perlakuan']==2])*selisih_p2*selisih_p2.T
A3=len(df['Perlakuan'][df['Perlakuan']==3])*selisih_p3*selisih_p3.T
```

```python
print("A1:")
print(A1)
print("\nA2:")
print(A2)
print("\nA3:")
print(A3)
```

```python
A=A1 + A2 + A3
print("\nA (Jumlah):")
print(A)
```

## D

```python
selisih_m1=m1-m_rata2_p1
selisih_m2=m2-m_rata2_p1
selisih_m3=m3-m_rata2_p2
selisih_m4=m4-m_rata2_p2
selisih_m5=m5-m_rata2_p2
selisih_m6=m6-m_rata2_p3
selisih_m7=m7-m_rata2_p3
```

```python
print("Selisih m1:")
print(selisih_m1)
print("Selisih m2:")
print(selisih_m2)
print("Selisih m3:")
print(selisih_m3)
print("Selisih m4:")
print(selisih_m4)
print("Selisih m5:")
print(selisih_m5)
print("Selisih m6:")
print(selisih_m6)
print("Selisih m7:")
print(selisih_m7)
```

```python
D= selisih_m1*selisih_m1.T + selisih_m2*selisih_m2.T + selisih_m3*selisih_m3.T + selisih_m4*selisih_m4.T + selisih_m5*selisih_m5.T + selisih_m6*selisih_m6.T + selisih_m7*selisih_m7.T
print("\nD (Jumlah):")
print(D)
```

## A+D

```python
AD= A + D
print("\nA+D:")
print(AD)
```

# SOAL 4

```python
detD=np.linalg.det(D)
detAD=np.linalg.det(AD)

print("\nDeterminant D:", detD)
print("Determinant A+D:", detAD)
```

```python
wilk=detD/detAD
print("\nWilks' Lambda:", wilk) 
```

```python
cricital_value = (((len(df))-df['Perlakuan'].nunique()-1)/(df['Perlakuan'].nunique()-1))*(1-np.sqrt(wilk))/(np.sqrt(wilk))
print("Critical Value:", cricital_value)
```

```python
from scipy import stats

f_table = stats.f.ppf(0.95, 4, 6)
print("F-table (α=0.05, df1=4, df2=6):", f_table)
```

```python
#uji wilks' lambda
if cricital_value > f_table:
    print("\nTolak H0: Ada perbedaan yang signifikan antara perlakuan.")
else:
    print("\nGagal Tolak H0: Tidak ada perbedaan yang signifikan antara perlakuan.")
```
