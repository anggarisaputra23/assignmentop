```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
```

```python
mean=5
stddev=1
size=100
upper_limit=10
lower_limit=1

rand=np.random.normal(mean, stddev, size)
```

```python
rand = np.clip(rand, lower_limit, upper_limit)
```

#ukuran pemusatan data

```python
rand_mean=np.mean(rand)
rand_median=np.median(rand)
rand_mode=statistics.mode(rand)
```

```python
sns.histplot(data=rand, bins=10, kde=True)
plt.axvline(x=rand_mean, color='red', linestyle='--', label='mean', linewidth=2)
plt.axvline(x=rand_median, color='blue', linestyle='dashdot', label='median', linewidth=2)
plt.axvline(x=rand_mode, color='black', linestyle='dotted', label='mode', linewidth=2)
plt.legend()
plt.show()
```
