```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
import pandas as pd
from scipy import stats
import math
```

```python
mean=5
stddev=1
size=100
upper_limit=10
lower_limit=1
bins=round(1+(3.3*(math.log(size))))

rand=np.random.normal(mean, stddev, size)
```

```python
rand = np.clip(rand, lower_limit, upper_limit)
```

```python
hist, bin_edges = np.histogram(rand, bins=bins)
# Create a DataFrame from the histogram data
df = pd.DataFrame({'Lower Bin Edge': bin_edges[:-1], 'Upper Bin Edge': bin_edges[1:], 'Frequency': hist})
print(df)
```

#ukuran pemusatan data

```python
#ukuran pemusatan
rand_mean=np.mean(rand)
rand_median=np.median(rand)
rand_mode=statistics.mode(rand)
q1=stats.scoreatpercentile(rand, 25)
q2=stats.scoreatpercentile(rand, 25)
q3=stats.scoreatpercentile(rand, 25)

#ukuran pesebaran
range=max(rand)-min(rand)
varians=np.var(rand)
std_dev=np.std(rand)
koefisien_varians=(std_dev/rand_mean)*100
```

```python
print("berikut ukuran pemusatan dari data")
print(f'mean = {rand_mean}')
print(f'median = {rand_median}')
print(f'modus = {rand_mode}')
print(f'quartile 1 = {q1}')
print(f'quartile 2 = {q2}')
print(f'quartile 3 = {q3}')

print("berikut ukuran pemusatan dari data")
print(f'range = {range}')
print(f'varians = {varians}')
print(f'standar deviasi = {std_dev}')
print(f'koefisien varians = {koefisien_varians:.2f}%')
```

```python
sns.histplot(data=rand, bins=bins, kde=True)
plt.axvline(x=rand_mean, color='red', linestyle='--', label='mean', linewidth=2)
plt.axvline(x=rand_median, color='blue', linestyle='dashdot', label='median', linewidth=2)
plt.legend()
plt.show()
```
