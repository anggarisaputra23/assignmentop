```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
from scipy import stats
import math
```

```python
mean=62
stddev=40
size=100
bins=round(1+(3.3*(math.log(size))))

rand=np.random.normal(mean, stddev, size)
```

```python
rand
```

```python
#ukuran pemusatan
rand_mean=np.mean(rand)
rand_median=np.median(rand)
rand_mode=statistics.mode(rand)
q1=stats.scoreatpercentile(rand, 25)
q2=stats.scoreatpercentile(rand, 50)
q3=stats.scoreatpercentile(rand, 75)

#ukuran pesebaran
range=max(rand)-min(rand)
varians=np.var(rand)
std_dev=np.std(rand)
koefisien_varians=(std_dev/rand_mean)*100
```

```python
print("berikut ukuran pemusatan dari data")
print(f'mean = {rand_mean}')
print(f'median = {rand_median}')
print(f'modus = {rand_mode}')
print(f'quartile 1 = {q1}')
print(f'quartile 2 = {q2}')
print(f'quartile 3 = {q3}')

print("berikut ukuran pemusatan dari data")
print(f'range = {range}')
print(f'varians = {varians}')
print(f'standar deviasi = {std_dev}')
print(f'koefisien varians = {koefisien_varians:.2f}%')
```

```python
randdf= pd.DataFrame({'Berat Badan': rand})
randdf
sns.histplot(randdf, bins=bins, kde=True)
plt.axvline(x=rand_mean, color='red', linestyle='--', label='mean', linewidth=2)
plt.axvline(x=rand_median, color='blue', linestyle='dashdot', label='median', linewidth=2)
plt.axvline(x=rand_mode, color='green', linestyle='dashdot', label='mode', linewidth=2)
plt.axvline(x=q1, color='black', linestyle='dashdot', label='q1', linewidth=2)
plt.axvline(x=q3, color='orange', linestyle='dashdot', label='q3', linewidth=2)
plt.legend()
plt.show()
```
