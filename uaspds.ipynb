# HYPOTHESIS TEST OF TWO POPULATION

```python
# Calculate the sample statistics
mean1 = df_laki['Berat Badan'].mean()
mean2 = df_perempuan['Berat Badan'].mean()
std1 = df_laki['Berat Badan'].std(ddof=0)
std2 = df_perempuan['Berat Badan'].std(ddof=0)
n1 = len(df_laki['Berat Badan'])
n2 = len(df_perempuan['Berat Badan'])

# Perform the t-test
t_stat, p_value = ttest_ind_from_stats(mean1, std1, n1, mean2, std2, n2)

print(f"t-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.4f}")
```
