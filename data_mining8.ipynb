```python
import numpy as np

# Data aktual dan prediksi dari tabel
aktual = np.array([26, 22, 24, 20, 27, 28, 24, 22, 27, 22])
prediksi = np.array([26.75, 21.22, 25.17, 23.59, 25.96, 26.75, 22.01, 23.59, 24.38, 22.8])
```

```python
import pandas as pd

df = pd.DataFrame({
    'Aktual': aktual,
    'Prediksi': prediksi
})
df
```

```python
df['Selisih'] = df['Aktual'] - df['Prediksi']
df['Selisih_kuadrat'] = df['Selisih'] ** 2
df['Selisih_absolut'] = np.abs(df['Selisih'])
df
```

```python

mse = np.mean(df['Selisih_absolut'])
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
rmse = np.sqrt(np.sqrt(np.mean(df['Selisih_kuadrat'])))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
