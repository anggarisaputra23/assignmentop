```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
import pandas as pd
from scipy import stats
import math
```

```python
# Define the discrete distribution
values = [1, 2, 3]
probabilities = [0.2, 0.5, 0.3]

# Number of random variables to generate
num_variables = 13984

# Generate random discrete variables
random_variables = np.random.choice(values, size=num_variables, p=probabilities)
df = pd.DataFrame({'(ordinal) e_wallet_user': random_variables})
```

```python
df['kategori_usia'] = df['(ordinal) e_wallet_user'].map({
    1: 'anak-anak',
    2: 'remaja',
    3: 'dewasa'
})
```

```python
sns.countplot(data=df, x='kategori_usia', hue='(ordinal) e_wallet_user', order=['anak-anak','remaja','dewasa'], palette='Set2')
plt.xlabel('kategori pengguna')
```
