```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
from scipy import stats
import math
```

```python
# Data for Perkotaan
data_perkotaan = {
    'Rumah Tangga': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Perkotaan': [110, 102, 98, 107, 120, 125, 115, 95, 122, 112]
}

# Data for Perdesaan
data_perdesaan = {
    'Rumah Tangga': ['P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y'],
    'Perdesaan': [54, 45, 60, 55, 75, 72, 62, 65, 85, 35]
}

# Create DataFrames
df_perkotaan = pd.DataFrame(data_perkotaan)
df_perdesaan = pd.DataFrame(data_perdesaan)

print("DataFrame Perkotaan:")
print(df_perkotaan)

print("\nDataFrame Perdesaan:")
print(df_perdesaan)
```

```python
sns.barplot(x='Rumah Tangga', y='Perkotaan', data=df_perkotaan)
plt.title('Data Perkotaan')
```

```python
# Combine the data into a single DataFrame for plotting
combined_data = pd.concat([
    df_perkotaan.rename(columns={'Perkotaan': 'Value'}).assign(Category='Perkotaan'),
    df_perdesaan.rename(columns={'Perdesaan': 'Value'}).assign(Category='Perdesaan')
])

# Create a bar plot
sns.barplot(x='Rumah Tangga', y='Value', hue='Category', data=combined_data)
plt.title('Perbandingan Konsumsi Perkotaan and Perdesaan')
plt.ylabel('Konsumsi')
plt.xlabel('Rumah Tangga')
plt.legend(title='Konsumsi')
plt.show()
```

```python
# ukuran pemusatan
rand_mean = np.mean(df_perkotaan['Perkotaan'])
rand_median = np.median(df_perkotaan['Perkotaan'])
rand_mode = statistics.mode(df_perkotaan['Perkotaan'])
q1 = stats.scoreatpercentile(df_perkotaan['Perkotaan'], 25)
q2 = stats.scoreatpercentile(df_perkotaan['Perkotaan'], 50)
q3 = stats.scoreatpercentile(df_perkotaan['Perkotaan'], 75)

# ukuran pesebaran
range = max(df_perkotaan['Perkotaan']) - min(df_perkotaan['Perkotaan'])
varians = np.var(df_perkotaan['Perkotaan'], ddof=1)
std_dev = np.std(df_perkotaan['Perkotaan'], ddof=1)
koefisien_varians = (std_dev / rand_mean) * 100
```

```python
print("berikut ukuran pemusatan dan persebaran dari data perkotaan")
print(f'mean = {rand_mean}')
print(f'median = {rand_median}')
print(f'modus = {rand_mode}')
print(f'quartile 1 = {q1}')
print(f'quartile 2 = {q2}')
print(f'quartile 3 = {q3}')

print("berikut ukuran pemusatan dari data")
print(f'range = {range}')
print(f'varians = {varians}')
print(f'standar deviasi = {std_dev}')
print(f'koefisien varians = {koefisien_varians:.2f}%')
```

```python
# ukuran pemusatan
rand_meand = np.mean(df_perdesaan['Perdesaan'])
rand_mediand = np.median(df_perdesaan['Perdesaan'])
rand_moded = statistics.mode(df_perdesaan['Perdesaan'])
q1d = stats.scoreatpercentile(df_perdesaan['Perdesaan'], 25)
q2d = stats.scoreatpercentile(df_perdesaan['Perdesaan'], 50)
q3d = stats.scoreatpercentile(df_perdesaan['Perdesaan'], 75)

# ukuran pesebaran
ranged = max(df_perdesaan['Perdesaan']) - min(df_perdesaan['Perdesaan'])
variansd = np.var(df_perdesaan['Perdesaan'], ddof=1)
std_devd = np.std(df_perdesaan['Perdesaan'], ddof=1)
koefisien_variansd = (std_devd / rand_meand) * 100
```

```python
print("berikut ukuran pemusatan dari data perdesaan")
print(f'mean = {rand_meand}')
print(f'median = {rand_mediand}')
print(f'modus = {rand_moded}')
print(f'quartile 1 = {q1d}')
print(f'quartile 2 = {q2d}')
print(f'quartile 3 = {q3d}')

print("berikut ukuran pemusatan dari data")
print(f'range = {ranged}')
print(f'varians = {variansd}')
print(f'standar deviasi = {std_devd}')
print(f'koefisien varians = {koefisien_variansd:.2f}%')
```

```python
# Create a dictionary to store the results
results = {
    'Metric': ['Mean', 'Median', 'Mode', 'Q1', 'Q2', 'Q3', 'Range', 'Variance', 'Std Dev', 'Coeff of Variation'],
    'Perkotaan': [rand_mean, rand_median, rand_mode, q1, q2, q3, range, varians, std_dev, koefisien_varians],
    'Perdesaan': [rand_meand, rand_mediand, rand_moded, q1d, q2d, q3d, ranged, variansd, std_devd, koefisien_variansd]
}

# Create a DataFrame
df_results = pd.DataFrame(results)

# Display the DataFrame
df_results
```

```python
# Convert the results dictionary to a DataFrame for easier plotting
df_results_melted = df_results.melt(id_vars='Metric', var_name='Category', value_name='Value')

# Create a bar plot to compare the metrics
plt.figure(figsize=(12, 6))
sns.barplot(x='Metric', y='Value', hue='Category', data=df_results_melted)
plt.title('Comparison of Metrics Between Perkotaan and Perdesaan')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot with values annotated
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Metric', y='Value', hue='Category', data=df_results_melted)

# Annotate the bars with their values
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=10, color='black', 
                xytext=(0, 5), 
                textcoords='offset points')

plt.title('Comparison of Metrics Between Perkotaan and Perdesaan')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```
